```python
from sklearn.preprocessing import StandardScaler

class Dataset:
    def __init__(self, data_points, reference_labeled_points, label, val_frac=0.5, do_normalize=True):
        
        train_data = data_points[label]
        test_data = reference_labeled_points[label]
        
        std_scaler = StandardScaler().fit(np.concatenate((train_data, test_data)))
        print('Scaler mean: ', std_scaler.mean_, ' scale: ', std_scaler.scale_)
        train_data_scaled = std_scaler.fit_transform(train_data)
        test_data_scaled = std_scaler.fit_transform(test_data)
        print('Train data size ', train_data_scaled.shape)
        print('Original test data: \n', test_data[:2, :])
        print('\nScaled test data: \n', test_data_scaled[:2, :])
        print('\nUnscaled test data: \n', std_scaler.inverse_transform(test_data_scaled[:2, :]))
        #y_scaler.inverse_transform(predicted)

        self.train_yz = train_data_scaled[:, 1:]
        self.train_x = train_data_scaled[:,0]
        
        self.num_ref_points = test_data_scaled.shape[0]
        if val_frac > 1e-8:
            split_idx = int(number_of_reference_points*(1 - val_frac))

            self.test_yz, self.valid_yz = test_data_scaled[:split_idx, 1:],\
                                          test_data_scaled[split_idx:, 1:]
            self.test_x, self.valid_x = test_data_scaled[:split_idx, 0], \
                                        test_data_scaled[split_idx:, 0]
        else:
            self.test_yz, self.valid_yz = None, test_data_scaled[:, 1:]
            self.test_x, self.valid_x = None, test_data_scaled[:, 0]            
        
    def batches(self, batch_size):
        n_batches = len(self.train_x)//batch_size
        for ii in range(0, len(self.train_x), batch_size):
            yz = self.train_yz[ii:ii+batch_size, :]
            x = self.train_x[ii:ii+batch_size]
            
            yield yz, x
```

```python
# data
dataset = Dataset(deformed_layer_shuffle_points4pids[used_pids[0]],
                  reference_labeled_points, '1', val_frac=0.0, do_normalize=True)
print(dataset.valid_x.shape)
```

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout, BatchNormalization
from keras.regularizers import l2, l1, l1_l2
```

```python
from keras.callbacks import History, ModelCheckpoint, CSVLogger, EarlyStopping
from keras.optimizers import SGD, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

kernel_reg=None

def base_model(num_hidden_units, in_dim, out_dim, dropouts=[]):
    """
        Number of hidden units, except the output layer: list, tuple or int
    """
    model = Sequential()
    
    if isinstance(num_hidden_units, list) or isinstance(num_hidden_units, tuple):
        model.add(Dense(num_hidden_units[0], input_dim=in_dim, \
                        activation='relu', kernel_initializer='glorot_uniform'))
        if dropouts: model.add(Dropout(dropouts[0]))
        for indx, num_hunit in enumerate(num_hidden_units[1:]):
            model.add(Dense(num_hunit, activation='relu', kernel_initializer='glorot_uniform'))
            if len(num_hidden_units)-2 > 0 and indx < len(num_hidden_units)-2 and indx+1<len(dropouts):
                print('Add dropout index ', indx+1, ' with rate ', dropouts[indx+1])
                model.add(Dropout(dropouts[indx+1]))
                
        #model.add(BatchNormalization())
    else:        
        model.add(Dense(num_hidden_units, input_dim=in_dim, \
                        activation='relu', kernel_initializer='glorot_uniform'))
        
    model.add(Dense(out_dim, kernel_initializer='glorot_uniform'))
    
    model.compile(loss='mean_absolute_error', optimizer = Nadam())
    
    return model

#seed = 7
#np.random.seed(seed)
```

```python
# params
epochs = 500
batch_size = dataset.num_ref_points
num_hiddens = [20, 10]
dropouts = [0.3]
input_dim, output_dim = 2,1

# model
model = base_model(num_hiddens, input_dim, output_dim, dropouts)
model.summary()

# Note:'weights.best_ndam.hdf5' & 'training_log_nadam.log' for d10 d1, epochs 500, batch_size 50000
# Affix name for log files
affix = ''.join([ 'd%d_' % n for n in num_hiddens])
affix += ('epch%d_btchsz%d' % (epochs,batch_size))
affix += 'drop'.join(['_%0.1f' % d for d in dropouts])
best_params_file = os.path.join(main_folder, 'save_weights', 'weights.best_ndam_%s.hdf5' % affix)
print('Best params file: ', best_params_file)
log_file = os.path.join(main_folder, 'save_weights', 'training_log_nadam_%s.log' % affix)
print('Log file: ', log_file)
```

```python
from keras.callbacks import History, ModelCheckpoint, CSVLogger, EarlyStopping
from keras.optimizers import Nadam

# Define callback
checkpointer = ModelCheckpoint(filepath=best_params_file, verbose=1, save_best_only=True)
history = History()
csv_logger = CSVLogger(log_file)
early_stopping = EarlyStopping(monitor='val_loss', min_delta=0,
                              patience=10, verbose=0, mode='auto')

## TODO: Train the model
model.fit(dataset.train_yz, dataset.train_x, validation_data=(dataset.valid_yz, dataset.valid_x),
          epochs=epochs, batch_size=batch_size, verbose=2,
          callbacks=[history, checkpointer, csv_logger, early_stopping])

## TODO: Save the model as model.h5
#model.save('my_model.h5')
```

```python
import csv

def read_log_file(filename):
    epochs, loss, val_loss = [], [], []
    with open(filename, 'r') as cvsfile:    
        reader = csv.reader(cvsfile, delimiter=',')
        for count, row in enumerate(reader):
            if not row or not row[0].isnumeric(): continue
            #print('Row ', row)
            epochs.append(row[0])
            loss.append(row[1])
            val_loss.append(row[-1])
    return epochs, loss, val_loss
    
```

```python
%matplotlib inline

num_hiddens = [10, 50, 100]#, 150, 200]
legend = [str(nh) for nh in num_hiddens]

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
min_val_loss = []
for nh in num_hiddens:
    #print('nh ', nh, ' epochs ', epochs, ' batch_size ', batch_size)
    affix = ('d%d_epch%d_btchsz%d' % (nh, epochs,batch_size))
    log_file = os.path.join(main_folder, 'save_weights', 'training_log_nadam_%s.log' % affix)
    #print(log_file)
    nb_epochs, loss, val_loss = read_log_file(log_file)
    min_val_loss.append(min(val_loss))
    ax1.plot(nb_epochs, val_loss)
    ax2.plot(nb_epochs, loss)
ax1.set_title('Regressor Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Val Loss')
ax1.legend(legend, loc='upper right')

ax2.set_title('Regressor Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend(legend, loc='upper right')
plt.tight_layout(pad=1, w_pad=1, h_pad=1.2)
print('Min val loss ', min_val_loss)
```

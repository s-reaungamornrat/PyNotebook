```python
import os
import os.path
import itertools
import pandas as pd

def get_meta_data(dirpath):
    dircomp = dirpath.split('_')
    dirpath, file = os.path.split(dirpath)
    return dircomp[-1].split('\\')[0], file

def get_region(filename):
    if 'differece' in filename:
        return 'Descriptiveness difference'
    elif 'choriod' in filename:
        return 'Choriod'
    elif 'retina' in filename:
        return 'Retina'
    elif 'homogenous' in filename:
        return 'Homogenous'
    elif 'match' in filename:
        return 'Match points'

filepath = 'C:/Users/Ja/Dropbox/data'
#def read_data_frames(file_path)

feature_ranks = dict()
for (dirpath, dirnames, filenames) in os.walk(filepath):
    pid = None
    multiple_dataframes = []
    for fidx in range(0,len(filenames)):
        filename = filenames[fidx]
        f = os.path.join(dirpath, filename)
        meta = get_meta_data(dirpath)
        current_pid = '_'.join(meta)
        data = pd.read_csv(f)
        region = get_region(f)
        data.columns = pd.MultiIndex.from_product([[region], data.columns])
        #print('{}\n\t{}'.format(region, data.columns))
        if not fidx: 
            pid = current_pid
            multiple_dataframes.append(data)
        elif fidx == len(filenames)-1:
            multiple_dataframes.append(data)
            if pid not in feature_ranks: 
                feature_ranks[pid] = pd.concat(multiple_dataframes, axis = 1)
            pid == current_pid 
        else:
            multiple_dataframes.append(data)
        
line = 'C:/Users/Ja/Dropbox/data\20170721_cc0001\left';
get_meta_data(line)

feature_ranks['cc0001_left']
```

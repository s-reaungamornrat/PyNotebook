```python
import json
import os
import random
import pandas as pd
import pprint
pd.set_option('display.max_colwidth', -1)
```

```python
def write_json_file(filename, data):
    with open(filename, 'w') as fp:
        json.dump(data, fp, indent=4, separators=(',', ':'), sort_keys=False)
```

```python
fixed_folder = './examples/20170830_figs'
fixed_files = os.listdir(fixed_folder)
moving_folder = './examples/20180115_figs'
moving_files = os.listdir(moving_folder)
transformed_img_folder = './examples/results_images'
transformed_img_files = os.listdir(transformed_img_folder)
video_folder = './examples/movies'
video_files = os.listdir(video_folder)
fixed_files
```

### Generate Results

```python
df = pd.read_csv('AAPL.csv', header =0)
uids = df['measure_uid'].tolist()
len(uids)
uids
```

```python
results=dict()
results['experiment_title'] = 'An Experiment' 
results['measurements'] = 'C:/D/dev/PyNotebook/report/AAPL.csv' 
for rid, uid in enumerate(uids):
    
    fixed_img_info = dict()
    fixed_img_info['header'] = 'Fixed Image'
    #fixed_img_info['column_number'] = 0
    fixed_img_info['type'] = 'image'
    fixed_img_info['path'] = os.path.join(fixed_folder,random.choice(fixed_files))
    
    moving_img_info = dict()
    moving_img_info['header'] = 'Moving Image'
    #moving_img_info['column_number'] = 1
    moving_img_info['type'] = 'image'
    moving_img_info['path'] = os.path.join(moving_folder,random.choice(moving_files))
    
    result_img_info = dict()
    result_img_info['header'] = 'Result Image'
    #result_img_info['column_number'] = 2
    result_img_info['type'] = 'image'
    result_img_info['path'] = os.path.join(transformed_img_folder,random.choice(transformed_img_files))
    
    video_img_info = dict()
    video_img_info['header'] = 'Registration'
    #video_img_info['column_number'] = 3
    video_img_info['type'] = 'video'
    video_img_info['path'] = os.path.join(video_folder, random.choice(video_files))
    
    result = dict()
    result['fixed'] = fixed_img_info
    result['moving'] = moving_img_info
    result['measure_uid'] = uid
    result['result_img'] = result_img_info
    result['video'] = video_img_info
    
    results[rid] = result
pprint.pprint(results)
```

```python
json_filename = 'C:/D/dev/PyNotebook/report/sample_results/results.json'
write_json_file(json_filename, results)
```

```python
def form_html_tag(item_type, path, video_type='mp4'):
    tag = None
    if item_type =='image':
        tag = '<img src=\"%s\" alt=\"\" border=3 class=\"img-icon\"></img>' % path
    elif item_type == 'video':
        tag = '<video controls><source src=\"%s\" type=\"video/%s\"></video>' % (path, video_type)
    return tag
        
print(form_html_tag('image', 'C:/D/dev/PyNotebook/reinforcement-learning/backup_diagram_v_pi.png'))
print(form_html_tag('video', 'C:/D/presentation/movies/movie.mp4'))
```

```python
def result_dataframe(data):

    result_dict = dict()
    result_dict['measure_uid'] = []

    for rid, result_collections in data.items():
        #print(result_collections)
        if not isinstance(result_collections, dict): continue
        for iid, result in result_collections.items():
            # Going into the dict describing each result containing description of fixed, moving, transformed images with measures
            if not isinstance(result, dict): 
                result_dict['measure_uid'].append(result)
                continue
                
            header, item_type, path = None, None, None
            for key, item in result.items():
                # Going into the innermost dict describing each item such as fixed image
                if key == 'header':
                    header = item
                    if item not in result_dict: # add additional table-column headers
                        result_dict[item] = []
                elif key == 'type':
                    item_type = item
                elif key == 'path':
                    path = item
            result_dict[header].append(form_html_tag(item_type, path))
            
    return pd.DataFrame(result_dict)

# table_columns = None
# result_dict = dict()
# result_dict['measure_uid'] = []

# for rid, result_collections in data.items():
#     print(result_collections)
#     if not isinstance(result_collections, dict): continue
#     table_columns = [None,]* (len(result_collections)+len(measure_columns))
#     for iid, result in result_collections.items():
#         # Going into the dict describing each result containing description of fixed, moving, transformed images with measures
#         if not isinstance(result, dict): 
#             result_dict['measure_uid'].append(result)
#             continue
            
#         header, item_type, path = None, None, None
#         for key, item in result.items():
#             # Going into the innermost dict describing each item such as fixed image
#             if key == 'header':
#                 header = item[1]
#                 if item[1] not in table_columns: # add additional table-column headers
#                     table_columns[item[0]] = item[1]
#                     if item[1] not in result_dict:
#                         result_dict[item[1]] = []
#             elif key == 'type':
#                 item_type = item
#             elif key == 'path':
#                 path = item
#         result_dict[header].append(form_html_tag(item_type, path))
```

```python
def generate_html(experiment_title, html_file, dataframe):
    table = dataframe.to_html(notebook=False, escape=False, justify='left', col_space=20,
          max_rows=1000, max_cols=1000).replace('<table border="1" class="dataframe">','<table class="sortable">')

    html_string = '''
<html>
    <head>
        <meta name="viewport" content="width=device-width, initial-scale=1">
        <link rel="stylesheet" href="./styles/staticpages.css" type="text/css">
        <style></style>
        <script src="./js_lib/sorttable.js"></script>

    </head>
    <body>
        <div class="banner-image"></div>


        <h1>'''+experiment_title+'''</h1>

        <div>
        ''' + table + '''
        </div>

    </body>
</html>'''     

    f = open(html_file,'w')
    f.write(html_string)
    f.close()
    
    return html_string
```
